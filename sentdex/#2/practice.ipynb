```python
# get all of the values from df table where value for "Low" is 0.0
issue_df = df[df["Low"] == 0]
issue_df.tail()
```

```python
issue_df["State"].unique()
```

```python
import numpy as np

act_min_wage.replace(0.0, np.NaN).dropna(axis=1).corr().head()
```

```python
# replace 0.0 with NaN and drop those columns from the act_min_wage table
# hint: the act_min_wage table has the states as the column titles. we are effectively removing the columns from being looked at
min_wage_corr = act_min_wage.replace(0.0, np.NaN).dropna(axis=1).corr()

# check to make sure all of the values are no longer in the table
for problem in issue_df["State"].unique():
    if problem in min_wage_corr.columns:
        print("we're missing something here...")
```
